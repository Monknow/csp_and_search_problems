Little Timmy has received a 3x3 sliding puzzle as a gift and wants to solve it to surprise his dad. The puzzle consists of 8 numbered tiles (from 1 to 8) and one empty slot, represented by the number 0. Timmy can move any tile adjacent to the empty slot into that slot, effectively "sliding" the empty space around the grid.

His goal is to reach the solved state, where the tiles are arranged in numerical order from the top-left to the bottom-right, with the empty slot in the first (top-left) position:

**INPUT FORMAT (search.in):**

The first, and only line of input, contains 9 integers separated by spaces. These represent the tiles in the grid starting from the top-left corner and moving row by row to the bottom-right corner.

**OUTPUT FORMAT (file search.out):**

Output the shortest number of steps needed to solve the puzzle. If the puzzle cannot be solved from the given configuration, print "None".

**SAMPLE INPUT:**

```
0 1 2 3 4 5 6 7 8
```

**SAMPLE OUTPUT:**

```
26
```

# Solution

````python
from __future__ import annotations
from generic_search import node_to_path, astar, Node
from typing import List, Optional
from math import floor
````

````python
goal_state =   [0, 1, 2, 
                3, 4, 5, 
                6, 7, 8]

possible_swaps = {
    0: [1, 3],
    1: [0, 2, 4],
    2: [1, 5],
    3: [0, 4, 6],
    4: [1, 3, 5, 7],
    5: [2, 4, 8],
    6: [3, 7],
    7: [6, 4, 8],
    8: [7, 5]
}
    


class Puzzle:
    def __init__(self, state: List[int]):
        self.state = state
    def goal_test(self) -> bool:
        return self.state == goal_state
    def successors(self) -> List[Puzzle]:
        puzzles: List[Puzzle] = []

        slot = self.state.index(0)

        for b in possible_swaps[slot]:
            new_state = self.state.copy()

            new_state[slot], new_state[b] = new_state[b], new_state[slot]

            puzzles.append(Puzzle(new_state))
            
        return puzzles
    def get_cost() -> float:
        return 1    


    def get_heuristic(self) -> float:
        total = 0

        for a, i in enumerate(self.state):
            x_i = i % 3
            y_i = floor(i / 3)

            x_f = a % 3
            y_f = floor(a / 3)

            distance = abs(x_i - x_f) + abs(y_i - y_f)
            total += distance

        return total
    
    def __hash__(self) -> int:
        return int("".join([str(a) for a in self.state]))
    def __eq__(self, value):
        return self.state == value.state


if __name__ == "__main__":

    with open("search.in", "r", encoding="utf-8") as file:
        inputs = [[int(x) for x in line.split()] for line in file]  

    results = []

    for initial in inputs:
        start: Puzzle = Puzzle(initial)
        solution: Optional[Node[Puzzle]] = astar(start, Puzzle.goal_test, Puzzle.successors, Puzzle.get_heuristic, Puzzle.get_cost)

        if solution:
            path = node_to_path(solution)
            results.append(str(len(path) - 1))
        else:
            results.append("None")

    with open("search.out", "w", encoding="utf-8") as out_file:
        out_file.write("\n".join(results) + "\n")
````
